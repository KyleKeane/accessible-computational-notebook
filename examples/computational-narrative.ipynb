## Setup

The next cell is an **initialization cell** (its label says so). After reopening this notebook, run *Cell > Run Initialization Cells* to restore the session in one command.

```python
monthly_sales = [1200, 1350, 980, 1500, 1720, 1650]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
```

## Exploration

Try **Ctrl+E** (Quick Evaluate) and type `max(monthly_sales)` to hear the answer without adding a cell. Or select part of the next cell and press **F9**.

```python
total = sum(monthly_sales)
average = total / len(monthly_sales)
round(average, 1)
```

```python
# _ is the last result; Out[n] is the result of execution n.
f"Average sales were {_} per month, {total} in total."
```

## Findings

The best and worst months, computed and spoken:

```python
best = months[monthly_sales.index(max(monthly_sales))]
worst = months[monthly_sales.index(min(monthly_sales))]
print(f"Best month: {best}. Slowest month: {worst}.")
```
